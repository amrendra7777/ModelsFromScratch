```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
x = np.random.randn(100)
```

```python
noise = np.random.randn(100)*0.1
```

```python
y = 2*x + 3 + noise
```

```python
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Synthetic Data: y=2x+3+noise")
```

```python
w = 0.0
b = 0.0
lr = 0.1
```

```python
def predict(x, w, b):
    return w*x + b
```

```python
y_pred = predict(x, w, b)
plt.scatter(x, y_pred)
```

```python
def compute_loss(y, y_pred):
    mse = 0
    for i in range(0, 100):
        mse += (y[i]-y_pred[i])**2
    mse = mse/100
    return mse
```

```python
mse = compute_loss(y, y_pred)
print(mse)
```

```python
def update_parameters(x, y, y_pred, w, b, lr):
    n = len(x)

    dw = (-2 / n) * np.sum((y - y_pred) * x)
    db = (-2 / n) * np.sum((y - y_pred))

    w = w-lr*dw
    b = b-lr*db

    return w, b
```

```python
for epoch in range(100):
    y_pred = predict(x, w, b)
    loss = compute_loss(y, y_pred)
    w, b = update_parameters(x, y, y_pred, w, b, lr)

    if epoch % 10 == 0:
        print("Epoch", epoch, ": Loss =", round(loss, 4), ", w =", round(w, 4), ", b =", round(b, 4))
```

```python
plt.scatter(x, y_pred)
```

```python
y_pred_final = predict(x, w, b)
plt.scatter(x, y, label="Actual data")
plt.plot(x, y_pred_final, color='red', label="Predicted line")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Learned Regression Line vs Data")
plt.show()
```
